# Print the top 5 teams which has the highest Winning percentage

```python
df['GamesWon']=df1.GamesWon.astype(str).astype(int)
df['GamesPlayed']=df1.GamesPlayed.astype(str).astype(int) 

def per_win(df):
    try:
        Per_win = df['GamesWon']/df['GamesPlayed']*100
        return Per_win
    except ZeroDivisionError:
        print("Replace Divide by Zero to NaN")

#df_add=df.copy()
df['Winning Percent']=df.apply(per_win,axis=1)
df['Winning Percent'].fillna(0)
```
